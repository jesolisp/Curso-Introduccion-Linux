## Paquetería de oficina

* [AbiWord](https://www.abisource.com/) — Procesador de texto que proporciona una versión ligera a LibreOffice Writer y OpenOffice Writer. Este programa tiene soporte a diferentes tipos de documentos, por ejemplo `.odf`, `.doc`, `.docx`, `.rtf`.

* [Apache OpenOffice](https://www.openoffice.org/why/). — Paquete de software de oficina de código abierto que contiene un procesador de textos, hojas de cálculo, presentaciones, gráficos, bases de datos y más. Tiene soporte para muchos idiomas y funciona en todas las computadoras comunes. Se puede descargar y utilizar de forma gratuita.

* [LibreOffice](https://www.libreoffice.org/) — Suite de código abierto gratuita multiplataforma (Windows, Mac y Linux) que proporciona un procesador de textos (Writer), hojas de cálculo (Calc), presentaciones (Impress), gráficos (Draw), bases de datos (Base).

* [WPS Office](https://www.wps.com/linux) — Alternativa patentada para Microsoft Office con una interfaz de usuario moderna. Permite la transferencia de archivos entre dispositivos y la copia de seguridad en la nube. Esta suite proporciona un procesador de textos (WPS Writer),  presentaciones (WPS Presentation), hojas de cálculo (WPS Spreadsheets).

* [Calligra](https://calligra.org/) — Bifurcación desarrollada de KOffice. Pertenece a la suite ofimática de KDE. Proporciona también la mayoría de las funciones de OpenOffice y también tiene versiones para teléfonos inteligentes (Calligra Mobile) y tabletas (Calligra Active).

* [OnlyOffice](https://www.onlyoffice.com/) — Paquete de Office que combina editores de texto, hojas de cálculo y presentaciones.

* [SoftMaker Office](https://www.freeoffice.com/) — Suite de oficina propietaria completa, confiable, ultrarrápida y compatible con Microsoft Office con un procesador de textos, hoja de cálculo y software de gráficos de presentación.

* [Yozo Office](https://www.yozosoft.com/product-officelinux.html) — Suite ofimática patentada, compatible con MS Office.


## Editores de imágenes

* [AzPainter](https://gitlab.com/azelpg/azpainter) — Software de pintura para dibujo ilustrativo.

* [CorePaint](https://cubocore.org/) — Herramienta de pintura de mapa de bits simple para producir gráficos muy simples. Parte de C-Suite.

* [Deepin Draw](https://github.com/linuxdeepin/deepin-draw) — Herramienta de dibujo ligera para escritorio Deepin.

* [Drawpile](https://drawpile.net/)  — Software de dibujo colaborativo que permite a varios usuarios dibujar en el mismo lienzo simultáneamente.

* [GIMP](https://www.gimp.org/) — Suite de edición de imágenes en la línea de los editores propietarios como Adobe Photoshop. 

* [KolourPaint](http://kolourpaint.org/) — Editor de gráficos rasterizados gratuito para KDE, similar a la aplicación Paint de Microsoft antes de Windows 7, pero con algunas funciones adicionales, como soporte para transparencia. 

* [Krita](https://krita.org/) — Software de pintura e ilustración digital incluido basado en la plataforma KDE.

* [Pinta](https://pinta-project.com/) — Software de dibujo y edición inspirado en Paint.NET. Su objetivo es proporcionar una alternativa simplificada a GIMP para usuarios ocasionales.

## Editores de audio

* [Audacity](https://www.audacityteam.org/)  — Software que le permite manipular formas de onda de audio digital.

* [SoundRecorder](https://wiki.gnome.org/Apps/SoundRecorder)  — Aplicación que permite grabar y reproducir archivos de sonido .flac, .ogg (audio OGG o .oga) y .wav.

* [Gnome Wave Cleaner](http://gwc.sourceforge.net/)  —  Editor de audio digital para eliminar el ruido, eliminar el silbido y amplificar los archivos de audio.

* [Kwave](http://kwave.sourceforge.net/) — Editor de sonido para KDE.

* [ocenaudio](https://www.ocenaudio.com/es/) — Editor de audio patentado multiplataforma, fácil de usar, rápido y funcional.

* [Snd](https://ccrma.stanford.edu/software/snd/) — Editor de sonido avanzado inspirado libremente en Emacs.

* [Barrer](http://www.metadecks.org/software/sweep/) — Editor de audio y herramienta de reproducción en vivo.

* [WaveSurfer](https://www.speech.kth.se/wavesurfer/) — Herramienta para visualización y manipulación de sonido. Las aplicaciones típicas son el análisis de voz/sonido y la anotación/transcripción de sonido.


## Editores de video

* [Avidemux](http://fixounet.free.fr/avidemux/) — Editor de video gratuito diseñado para tareas simples de corte, filtrado y codificación.

* [Blender](https://www.blender.org/) — Software de creación de gráficos 3D totalmente integrado con un editor de video.

* [Cinelerra](http://cinelerra-cv.wikidot.com/) — Entorno avanzado de composición y edición de video.

* [Flowblade](https://jliljebl.github.io/flowblade/) — Editor de video multipista para Linux,  para brindar una experiencia de edición rápida y sólida.

* [Kdenlive](https://kdenlive.org/) — Editor de video diseñado para trabajos de edición, desde básicos a semiprofesionales.

* [Lightworks](https://www.lwks.com/) — Sistema de edición propietario profesional para editar y masterizar video digital en diversos formatos.

* [LosslessCut](https://mifi.no/losslesscut/) — Herramienta para recortar o cortar archivos de video y audio sin pérdidas.

* [Natron](https://natrongithub.github.io/) — Software de código abierto. Se basa en gráficos de nodos. Similar a softwares como Adobe After Effects y Nuke de The Foundry.

* [Olive](https://www.olivevideoeditor.org/) — Editor de video que proporciona una alternativa  a software de edición de video profesional de gama alta.

* [OpenShot](https://www.openshot.org/) — Editor de video basado en el marco MLT.

* [Pitivi](https://www.pitivi.org/) — Editor de video intuitivo e integrado en el escritorio GNOME.

* [Shotcut](https://www.shotcut.org/) — Editor de video gratuito, de código abierto y multiplataforma.


## Reproductores de audio 

* [Clementine](https://www.clementine-player.org/) — Clon de Amarok, portado a Qt5.

* [Exaile](https://www.exaile.org/) — Clon GTK de Amarok.

* [GNOME Music](https://wiki.gnome.org/Apps/Music) — Software de reproducción de música de GNOME. 

* [Melody](https://anufrij.org/melodia/) — Reproductor de música para reproducir archivos locales, estaciones de radio en línea y CD de audio.

* [Rhythmbox](https://wiki.gnome.org/Apps/Rhythmbox) — Clon GTK de iTunes, usado por defecto en GNOME.

* [Sayonara](https://sayonara-player.com/) — Reproductor de audio minimalista y ligero para Linux escrito en C ++, utiliza el marco Qt.

* [Tauon Music Box](https://tauonmusicbox.rocks/) — Reproductor de música optimizado para la reproducción de su colección de música.

## Reproductores de video

* [MPlayer](https://mplayerhq.hu/design7/news.html) Reproductor de video que admite diversos formatos de video y audio.

* [mpv](https://mpv.io/) — Reproductor de películas basado en MPlayer y mplayer2.

* [Reproductor multimedia VLC](https://www.videolan.org/vlc/): Software que permite reproducir videos de alta definición sin problemas.
